```python
import pandas as pd
import numpy as np

import requests
import joblib
```

```python
important_features = joblib.load('data/cleaned/list_col_to_keep_from_train_application_final.joblib')
train = pd.read_csv("data/source/application_train.csv", usecols=important_features+['TARGET','SK_ID_CURR'])
test = pd.read_csv("data/source/application_test.csv",  usecols= important_features+['SK_ID_CURR'] )
```

```python
train.to_csv("data/source/application_train_clean.csv",header = True,index=False)
test.to_csv("data/source/application_test_clean.csv",header = True,index=False)
```

```python
import joblib
HistGB_pipe = joblib.load( "data/model-save/HistGradientBoostingClassifier_model.sav")
lgbm = joblib.load("data/model-save/lightGBMClassifier_model.sav")
```

```python
from sklearn.model_selection import train_test_split

X = train.drop(columns=['TARGET','SK_ID_CURR'])
y = train['TARGET']

## inf
X[X==np.inf] = np.nan

#X = X[X['CODE_GENDER'] != 'XNA']

for bin_feature in ['CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY']:
    X[bin_feature], uniques = pd.factorize(X[bin_feature])
X['INCOME_CREDIT_PERC'] = X['AMT_INCOME_TOTAL'] / X['AMT_CREDIT']
X['time_to_repay'] = X['AMT_CREDIT'] / X['AMT_ANNUITY']
X['ANNUITY_INCOME_PERC'] = X['AMT_ANNUITY'] / X['AMT_INCOME_TOTAL']

random_seed = 971

# Le dataset est grand on peut diviser le donn√©es en train test valiation 
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=random_seed,stratify=y) #~~100000 samples
X_test, X_validation, y_test, y_validation = train_test_split( X_test, y_test, test_size=0.5, random_state=random_seed,stratify=y_test) 
```

```python
from sklearn.inspection import permutation_importance

feature_names = [f"feature {i}" for i in range(X.shape[1])]

result = permutation_importance(
    HistGB_pipe, X_test, y_test, n_repeats=5, random_state=42, n_jobs=2
)
#print(f"Elapsed time to compute the importances: {elapsed_time:.3f} seconds")

forest_importances = pd.Series(result.importances_mean, index=X.columns)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=result.importances_std, ax=ax)
ax.set_title("Feature importances using permutation on full model")
ax.set_ylabel("Mean accuracy decrease")
fig.tight_layout()
plt.show()
```
